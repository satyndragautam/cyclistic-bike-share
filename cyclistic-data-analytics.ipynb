We have so far refined datasets/dataframes as we require for analysis. Now we will combine them together for processing/data cleaning for further analysis. 

```r
#Combining the datasets together

all_months_rides <- bind_rows(apr_2020, may_2020, jun_2020, jul_2020, aug_2020, sep_2020, oct_2020, nov_2020, dec_2020, jan_2021,
                            feb_2021, mar_2021, apr_2021)

glimpse(all_months_rides)
```

We notice that dates are in the character data type. We need to convert them in the right format to be precisely consistent. 

```r
#changing the dates into dttm format

all_months_rides$started_at <- parse_date_time(all_months_rides$started_at,"%Y-%m-%d %H:%M:%S")

all_months_rides$ended_at <- parse_date_time(all_months_rides$ended_at,"%Y-%m-%d %H:%M:%S")

glimpse(all_months_rides)
```

We are going to use this code for later requirements if needed to find the key insights... `format(strptime(all_months_rides$ended_at, format='%Y-%m-%d %H:%M:%S'), '%Y-%m-%d %I:%M:%S %p') & format(strptime(all_months_rides$started_at, format='%Y-%m-%d %H:%M:%S'), '%Y-%m-%d %I:%M:%S %p')`


To find out how long riders take time/ride, we will need to create another variable `ride_length`. 

```r
#mutating a new column for ride length

all_months_rides <- all_months_rides %>%
    mutate(ride_length_mins = round(difftime(ended_at, started_at, units = "mins"),0))

glimpse(all_months_rides)
```

In order to carry on with consistent datasets, we need to check out if the `ride_length` in right format. if not so, we will convert in the numeric value. 

```r
#changing the `drtn` type into numeric value type

all_months_rides$ride_length_mins <- as.numeric(all_months_rides$ride_length_mins)

glimpse(all_months_rides)
```

<a id="Process-5.3"></a>
### ▣ **5.3 Looking for the NA, duplicates and Unique values:**

Lets look for the duplicate values, null, and unique values for this data frame. 

```r
n_unique(all_months_rides$ride_id) #unique ride_id

sum(duplicated(all_months_rides)) #no duplicates in the overall data frame but we still need to check by ride_id

summary(all_months_rides) #getting summary

colSums(is.na(all_months_rides)) #checking the NA values in the variables
```

* What we notice while cleaning the data frame is ride_length_mins are in negative and it will be outliners for your analysis. We will also notice if we extract only duplicated values by ride_id, we will see that dates are messy and not accurate. To simply put, we will extract those values before analysis. 

```r
duplicate_ride_id <- all_months_rides[duplicated(all_months_rides$ride_id),] #checking the duplicated values by ride_id

glimpse(duplicate_ride_id)#glimpse of data structure which clears that there are messy and inaccurate values (209)

non_duplicated_rides <- all_months_rides[!duplicated(all_months_rides$ride_id),] #extracting duplicates and inaccurate data
    
glimpse(non_duplicated_rides)
summary(non_duplicated_rides)#checking out if the dates are still inaccurately recordered.
```

* We found out that there are outliners(negative ride_length and excessive positive length). We need to remove those values from the final data frame for analysis.

```r
#multiplying the 24*60 =1440 to determiine the correct ride_length since it is bike-share company

final_months_rides <- non_duplicated_rides[!(non_duplicated_rides$ride_length >= 1440 | non_duplicated_rides$ride_length <= 0),]

glimpse(final_months_rides)
summary(final_months_rides)
colSums(is.na(final_months_rides))
```

### **<span style = "color: #e8ab07">① Distribution of Riders and Average time spent by riders:</span>**
* **

* We will determine what percentage of casual riders and annual riders are. 

```r
str(final_months_rides)

distribution_riders <- final_months_rides %>%
    group_by(member_casual) %>%
    summarize(total = n()) %>%
    mutate(overall_total = sum(total)) %>%
    group_by(member_casual) %>%
    summarize(percent_total = total/overall_total) %>%
    ggplot(aes(x = "", y = percent_total, fill = member_casual)) +
    geom_col(color = "#46465e", size = 0.5) +
    coord_polar("y", start = 0) +
    geom_text(aes(label = scales :: percent(percent_total)), position = position_stack(vjust = 0.5), size = 10, fontface = "bold", color = "#46465e") +
    scale_fill_manual(values = c("#fae2a3", "#e8ab07"), name = "Types of riders", breaks = c("casual", "member"), labels = c("Casual Member", "Annual Member")) +
    labs(title = "Distribution of Riders", 
       subtitle = "What percentage of riders are using the Cyclistic?", 
       caption = paste0("Cyclistic historical data: ", min_date, " to ", max_date)) + 
    theme(panel.background = element_blank(),
        axis.line = element_blank(),
        axis.text = element_blank(),  
        axis.ticks = element_blank(),
        axis.title = element_blank(),
        plot.title = element_text(hjust = 0.5, size = 19, face = "bold", color = "#e8ab07"),
        plot.subtitle = element_text(hjust = 0.5, face = "bold", size = 13, color = "#46465e"),
        plot.caption = element_text(hjust = 5, size = 12, color = "#46465e", face = "bold"),
        legend.title = element_text(size = 16, face = "bold", color = "#46465e"),
        legend.text = element_text(size = 14, color = "#46465e"))

print(distribution_riders)
```

* **
**✒Insight 1:** As we can see that there are **<span style = "color:#e8ab07">59%</span>** of riders are carrying the annual membership while **<span style = "color:#e8ab07">41%</span>** are casual users who use full-day pass or single ride pass. 
* **

* Now we will figure out, on the average, how long a customer use the bike by rider type?

```r
average_time_spent <- final_months_rides %>%
    group_by(member_casual) %>%
    summarize(avg_time = mean(ride_length_mins)) %>%
    ggplot(aes(x = member_casual, y = avg_time)) +
    geom_col(fill = "#46465e") + 
    geom_text(aes(label = round(avg_time,0)), vjust = 2, size = 10, color = "#e8ab07", fontface = "bold") + 
    labs(title = "Average time spent by riders", subtitle = "On the average, casual riders spend more time than riders with membership",
        x = "Type of riders", y = "Average time spent(mins)", caption = paste0("Cyclistic historical data: ", min_date, " to ", max_date)) +
    theme(plot.title = element_text(hjust = 0.5, size = 19, color = "#e8ab07", face = "bold"), 
    plot.subtitle = element_text(hjust = 0.5, size = 16, color = "#46465e", face = "bold"), 
    plot.caption = element_text(size = 12, color = "#e8ab07", face = "bold"),
    axis.text.x = element_text(size = 14, color = "#e8ab07", face = "bold"),
    axis.text.y = element_text(size = 14, color = "#e8ab07", face = "bold"),
    axis.title.x = element_text(size = 18, color = "#46465e", face = "bold"), 
    axis.title.y = element_text(size = 18, color = "#46465e", face = "bold"),
    strip.text.x = element_text(size = 13), 
    strip.text.y = element_text(size = 13))

print(average_time_spent)
```

* **
**✒Insight 2:** It is pretty clear that the casual riders spend time using the bikes more than riders with annual membership. On average, Annual member spend **<span style = "color:#e8ab07">16 mins</span>** while casual riders spend **<span style = "color:#e8ab07">36 mins</span>**.

### **<span style = "color: #e8ab07">② Bikes usage and time spent with bikes by riders:</span>**

we will determine which bike has been utilized more?

```r
bike_usage_percentage <- final_months_rides %>%
    group_by(rideable_type) %>%
    summarize(total = n()) %>%
    mutate(overall_total = sum(total)) %>%
    group_by(rideable_type) %>%
    summarize(percent_total = total/overall_total) %>%
    ggplot(aes(x = 2, y = percent_total, fill = rideable_type)) +
    geom_col(color = "white") +
    coord_polar("y", start = 1) +
    geom_text(aes(label = scales :: percent(percent_total), fontface = "bold"), size = 9, color = "#e8ab07", position = position_stack(vjust = 0.5)) +
    scale_fill_manual(values = c("#8080a1","#46465e", "#53536f"), name = "Types of Bikes") +
    xlim(0.7, 2.5) + 
    labs(title = "Percentage of bikes used", 
       subtitle = "Which bike has been used the most among other types?", 
      caption = paste0("Cyclistic historical data: ", min_date, " to ", max_date)) + 
    theme(panel.background = element_blank(),
        axis.line = element_blank(),
        axis.text = element_blank(),  
        axis.ticks = element_blank(),
        axis.title = element_blank(),
        plot.title = element_text(size = 20, face = "bold", color = "#e8ab07", hjust = 0.5),
        plot.subtitle = element_text(face = "bold", size = 16, color = "#46465e", hjust = 0.5),
        plot.caption = element_text(size = 12, color = "#e8ab07", face = "bold", hjust = 3),
        legend.title = element_text(size=18, face = "bold", color = "#e8ab07"), 
        legend.text = element_text(size=14))

print(bike_usage_percentage)
```

* **
